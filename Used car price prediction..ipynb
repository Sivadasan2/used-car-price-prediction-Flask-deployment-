```python
df2=pd.get_dummies(df1,drop_first=True)
```

```python
df2.info()
```

```python
LR=LinearRegression()
```

```python
y=df2.loc[:,'Selling_Price']
```

```python
X=df2.drop('Selling_Price',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
LR_model=LR.fit(X_train,y_train)
```

```python
LR_model.coef_
```

```python
y_pred=LR_model.predict(X_test)
```

```python
err=mean_squared_error(y_test,y_pred)
print(np.sqrt(err))
```

```python
X_train.iloc[2]
```

```python
RFR=RandomForestRegressor()
```

```python
RFR_model=RFR.fit(X_train,y_train)
```

```python
y_pred1=RFR_model.predict(X_test)
```

```python
err=mean_squared_error(y_test,y_pred1)
print(np.sqrt(err))
```

```python
RFR_model.predict([list(X_train.iloc[2])])
```

```python
# Picling the file.

file = open('RFR_model.pickle','wb')
pickle.dump(RFR_model, file)
file.close()
```

```python
file = open('RFR_model.pickle', 'rb')
RF_model = pickle.load(file)
file.close()
    
def predict(a,model=RF_model):
    val=RF_model.predict(a)
    return round(float(val),2)
```
